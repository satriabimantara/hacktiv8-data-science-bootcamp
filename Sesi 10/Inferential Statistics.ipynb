## Random Variable
***
**Definisi** <br>
Suatu variabel yang kemungkinan nilainya adalah hasil numerik dari fenomena acak. <br>
 **Tipe** <br>
 1. Diskrit random variable <br>
 > Variable yang mungkin hanya mengambil sejumlah nilai berbeda yang dapat dihitung dan dapat dikuantifikasi <br>
 > Contoh, X adalah variabel acak sebagai angka yang muncul saat kita melempar sebuah dadu yang adil. Maka, X dapat berisi nilai (1,2,3,4,5,6) <br>
 > **Probability function** adalah probability distribution dari diskrit random variable yaitu daftar probabilitas yang terkait dengan masing-masing kemungkinan nilainya. <br>
 > P(X=xi) = pi didefinisikan sebagai variabel acak X dapat mengambil nilai k yang berbeda dengan probabilitas X=xi. pi harus memenuhi 0<pi<1 untuk setiap i, dan p1+p2+...+pk=1
 2. Continous random variable <br>
 > Continuous random variable adalah variable yang mengambil jumlah kemungkinan nilai yang tidak terbatas. Misalnya variabel acak X sebagai tinggi siswa di kelas. <br>
 > **Probability density function** adalah probability distribution dari continuous random variable. <br>
 > Sebagai contoh, variabel acak X adalah semua nilai yang berada dalam suatu interval bilangan real. Probabilitas bahwa X berada dalam himpunan hasil A disimbolkan dengan **P(A)**. P(A) adalah luas daerah diatas interval A dan dibawah kurva. Kurva yang merepresentasikan fungsi p(x) harus memenuhi:
 >> 1. kurva tidak memiliki nilai negatif p(x) > 0 untuk semua nilai x <br>
 >> 2. luas daerah total di bawah kurva = 1

# Probability Distribution Function (Continuous) 

## Uniform Distribution
***
**Probability distribution function dari contiuous uniform distribution** <br>



```python
from scipy.stats import uniform
```

loc, scale --> interval
size --> jumlah variasi acak

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n,loc=start,scale=width)
```

```python
# using seaborn distplot
ax = sns.distplot(
    data_uniform,
    kde=True,
    bins=100,
    color='skyblue',
    hist_kws={'linewidth':15,'alpha':1}
)
ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

```python
def plot_using_sns(data, kde=True):
    ax = sns.distplot(
        data,
        kde=kde,
        bins=100,
        color='skyblue',
        hist_kws={'linewidth':15,'alpha':1}
    )
    ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

# Confidence Intervals

## Definition
***
**Confidence ineterval** atau CI adalah estimasi yang dihitung dari data statistik yang diamati. CI mengukur **seberapa akurat Mean sebuah sampel** dapat mewakili **nilai mean populasi sesungguhnya**. CI juga diartikan sebagai rentang antara dua nilai dimana **suatu sampel Mean** tepat berada di tengah-tengahnya.

## Cara menghitung Confidence Intervals
***
Perhitungan CI melibatkan estimasi terbaik yang diperoleh dari **sample** dan **margin of errors**. <br>
Rumus dari CI: <br>
 <br>
*Keterangan* <br>
SE --> Standard Error <br>
z --> z_score <br>
1. Rumus SE dari population proportion

2. Rumus SE dari sample mean


### Contoh menghitung CI dari population proportion di Python

```python
import pandas as pd
import numpy as np
```

```python
df_heart = pd.read_csv('data/heart.csv')
```

```python
df_heart
```

```python
# replace value gender
df_heart['Sex1'] = df_heart['sex'].replace({1:"Male",0:'Female'})
```

```python
df_target_sex = df_heart[['target','Sex1']].dropna()
```

```python
df_target_sex
```

```python
# cari jumlah orang (pria, wanita) yang menderita dan tidak menderita penyakit jantung
pd.crosstab(df_target_sex.target,df_target_sex.Sex1)
```

#### Hitung proporsi penduduk wanita yang menderita penyakit jantung

```python
total_populasi_wanita = 86+226
prob_female_has_disease = 226/(total_populasi_wanita)
```

```python
prob_female_has_disease
```

#### Hitung Standard Error

```python
se_female = np.sqrt(prob_female_has_disease*(1-prob_female_has_disease)/total_populasi_wanita)
```

```python
se_female
```

#### Buat CI dengan z_score 95% z-score = 1.96

```python
z_score = 1.96
lower_limit_ci = prob_female_has_disease - (se_female*z_score)
upper_limit_ci = prob_female_has_disease + (se_female*z_score)
lower_limit_ci,upper_limit_ci
```

### Contoh menghitung CI dari sample mean

Kolom **chol** menunjukkan tingkat kolestorel

```python
df_heart.groupby('Sex1').agg({'chol':[np.mean,np.std,np.size]})
```

```python
col_mean_sample = 261.455128
col_std_sample = 64.466781
n_sample = 312
z = 1.96
```

```python
se_col_mean_sample = col_std_sample/np.sqrt(n_sample)
```

```python
lower_ci_col_mean_sample = col_mean_sample - z*se_col_mean_sample
upper_ci_col_mean_sample = col_mean_sample + z*se_col_mean_sample
```

```python
lower_ci_col_mean_sample,upper_ci_col_mean_sample
```

95% CI states that population mean of female colesterol level lies between 254,30 - 268,61

# Central limit theorem
***
Kalau kita sampling dari suatu distribusi apapun, maka mean dari sampel kita akan berdistribusi normal. <br>
Central limit theorem jika data kita ingin menghitung CI namun distribusi dari data tidak normal, maka lakukan sampling dulu pada datanya baru agar normal, baru lakukan perhitungan CI

## Simulasi Central Limit Theorem

```python
np.random.seed(42)

# make normal distribution data
normal_dict_us_male_mass_pounds = np.random.normal(loc=181,scale=24,size=6000)
normal_dict_us_female_mass_pounds = np.random.normal(loc=131,scale=22,size=6500)
all_mass = np.concatenate((normal_dict_us_male_mass_pounds,normal_dict_us_female_mass_pounds))
df = pd.DataFrame(data={'all_people_mass':all_mass})
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(all_mass, bins=100)
plt.show()
```

Distribusinya tidak normal, sehingga kita perlu lakukan sampling

```python
pop_mean_mass = df['all_people_mass'].mean()
pop_std_mass = df['all_people_mass'].std()
```

```python
# sampling 25 data set sebanyak 300x
sample_means = []
n = 25
for sample in range(0,300):
    sample_values = np.random.choice(a=df['all_people_mass'], size=n)
    
    # hitung sample mean
    sample_mean = np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
sns.distplot(sample_means)
```

```python
x = pd.Series(sample_means)
print('Skew = ',x.skew())
print('Mean = ',x.mean())
print('Median = ',x.median())
```

distribusinya sudah normal, baru hitung CI

```python
mean_of_sample_mean = np.mean(sample_means)
std_of_sample_mean = np.std(sample_means)
z_score = 1.96
```

```python
se_sample_means = pop_std_mass/np.sqrt(12500)
se_sample_means
```

```python
lower = mean_of_sample_mean-z_score*se_sample_means
upper = mean_of_sample_mean + z_score*se_sample_means
```

```python
lower,upper
```

**Kesimpulan** 95% CI states that the population mean of US people mass lies between 154,00 and 155,19
